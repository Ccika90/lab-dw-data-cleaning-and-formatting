## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
# Your code here# Standardize the Gender column
df
```

```python
# Standardize the Gender column
df['gender'] = df['gender'].str.strip().str.lower().map({
    'f': 'F', 'female': 'F', 'femal': 'F',
    'm': 'M', 'male': 'M'
})
df
```

```python
# Replace state abbreviations with full names
state_mapping = {
    "AZ": "Arizona",
    "Cali": "California",
    "WA": "Washington"
}
df['state'] = df['state'].map(state_mapping).fillna(df['state'])
df
```

```python
# Replace "Bachelors" with "Bachelor" in the Education column
df['education'] = df['education'].str.replace('Bachelors', 'Bachelor', regex=False)
df
```

```python
# Remove the '%' character from Customer Lifetime Value column
df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '', regex=False).astype(float)
df
```

```python
# Replace specific vehicle classes with "Luxury"
luxury_classes = ["Sports Car", "Luxury SUV", "Luxury Car"]
df['vehicle_class'] = df['vehicle_class'].replace(luxury_classes, 'Luxury')
df
```

Identify any columns with null or missing values. Identify how many null values each column has. You can use the `isnull()` function in pandas to find columns with null values.

Decide on a strategy for handling the null values. There are several options, including:

After formatting data types, as a last step, convert all the numeric variables to integers.

```python
# Your code here
# Function to fill null values based on column data type
def fill_nulls(df):
    for column in df.columns:
        if df[column].dtype == 'int64' or df[column].dtype == 'int32':  # Integer columns
            df[column].fillna(0, inplace=True)
        elif df[column].dtype == 'float64' or df[column].dtype == 'float32':  # Float columns
            df[column].fillna(0.0, inplace=True)
        elif df[column].dtype == 'object':  # String columns
            df[column].fillna('-', inplace=True)
    return df

# The reason for using this process is because I have in mind that the dataframe may be edited in the future to want to add more values to the places where there are no values

# Apply the function to the DataFrame
df = fill_nulls(df)

# Display the updated DataFrame
print(df)
```

```python
# Function to convert numeric values to integers, leaving non-convertible values and "Customer" column as is
def convert_to_int(value, column_name):
    if column_name == 'customer':
        # Leave values in the "Customer" column as is
        return value
    try:
        # Check if the value is numeric and does not contain letters
        if isinstance(value, (int, float)):
            return int(value)
        elif isinstance(value, str) and value.isdigit():
            return int(value)
        elif isinstance(value, str) and value.replace('.', '', 1).isdigit():
            return int(float(value))
        else:
            # Leave as is if it contains letters or cannot be converted
            return value
    except ValueError:
        # Leave as is if conversion fails
        return value

# Apply the function to the DataFrame, column by column
df_converted = df.apply(lambda col: col.map(lambda x: convert_to_int(x, col.name)))

# Display the updated DataFrame
print("DataFrame after converting numeric values to integers:")
print(df_converted)
```
